```python
import pandas as pd
```

```python
df2017 = pd.read_excel("./2017.xlsx")
df2017.iloc[:,:4]=df2017.iloc[:,:4].fillna(method="ffill")
df2017[["조회년도","조회월"]]=df2017[["조회년도","조회월"]].astype("int")
df2017
```

```python
df2018 = pd.read_excel("./2018.xlsx")
df2018.iloc[:,:4]=df2018.iloc[:,:4].fillna(method="ffill")
df2018[["조회년도","조회월"]]=df2018[["조회년도","조회월"]].astype("int")
df2018
```

```python
df2019 = pd.read_excel("./2019.xlsx")
df2019.iloc[:,:4]=df2019.iloc[:,:4].fillna(method="ffill")
df2019[["조회년도","조회월"]]=df2019[["조회년도","조회월"]].astype("int")
df2019
```

```python
df2020 = pd.read_excel("./2020.xlsx")
df2020.iloc[:,:4]=df2020.iloc[:,:4].fillna(method="ffill")
df2020[["조회년도","조회월"]]=df2020[["조회년도","조회월"]].astype("int")
df2020
```

```python
df2021 = pd.read_excel("./2021.xlsx")
df2021.iloc[:,:4]=df2021.iloc[:,:4].fillna(method="ffill")
df2021[["조회년도","조회월"]]=df2021[["조회년도","조회월"]].astype("int")
df2021
```

```python
result = pd.concat([df2017,df2018,df2019,df2020,df2021])
result.to_csv("result.csv", encoding="utf-8",index=False)
```
